###### Excercise: Create a program that creates a matrix using user input, displays it, displays its above-major diagonal elements, below-major diagonal elements, above-minor diagonal elements, below-minor diagonal elements, major diagonal elements, and minor diagonal elements.

 A = matrix, m = number of rows, n = number of columns

```python
def input_matrix():
    global A, m, n
    m = int(input("\nEnter number of rows: "))
    n = int(input("Enter number of columns: "))
    A = [[0 for col in range(n)] for row in range(m)]
    
    print("\nEnter your elements one-by-one below.")
    
    for i in range(m):
        for j in range(n):
            A[i][j] = int(input(f"Enter element at row {i+1} and column {j+1}: "))
            
    print("\nMatrix created.\n")
```

```python
def display_matrix(A, m, n):
    try:
        print("\nYour martix is:\n")
        for i in range(m):
            for j in range(n):
                print(A[i][j], end = "\t")
            print()
    except:
        print("You must first input a matrix.")
```

```python
def above_major(A, m, n):
    try:
        if m != n:
            print("\nTo display elements above the major diagonal, number of rows must equal number of columns.\n")
        else:
            print("\nYour martix elements above major diagonal are:\n")
            for i in range(m):
                for j in range(n):
                    if i < j:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```

```python
def below_major(A, m, n):
    try:
        if m != n:
            print("\nTo display elements below the major diagonal, number of rows must equal number of columns.\n")
        else:
            print("\nYour martix elements below major diagonal are:\n")
            for i in range(m):
                for j in range(n):
                    if i > j:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```

```python
def above_minor(A, m, n):
    try:
        if m != n:
            print("\nTo display elements above the minor diagonal, number of rows must equal number of columns.\n")
        else:
            print("\nYour martix elements above minor diagonal are:\n")
            for i in range(m):
                for j in range(n):
                    if i+j < n-1:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```

```python
def below_minor(A, m, n):
    try:
        if m != n:
            print("\nTo display elements below the minor diagonal, number of rows must equal number of columns.\n")
        else:
            print("\nYour martix elements below minor diagonal are:\n")
            for i in range(m):
                for j in range(n):
                    if i+j > n-1:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```

```python
def major_diagonal(A, m, n):
    try:
        if m != n:
            print("\nTo display the major diagonal elements, number of rows must equal number of columns.\n")
        else:
            print("\nThe major diagonal elements are:\n")
            for i in range(m):
                for j in range(n):
                    if i == j:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```

```python
def minor_diagonal(A, m, n):
    try:
        if m != n:
            print("\nTo display the minor diagonal elements, number of rows must equal number of columns.\n")
        else:
            print("\nThe minor diagonal elements are:\n")
            for i in range(m):
                for j in range(n):
                    if i+j == n-1:
                        print(A[i][j], end = "\t")
                    else:
                        print("\t", end = "")
                print()
    except:
        print("You must first input a matrix.")
```
