###### Excercise: Create a program that asks the user to input a list of integers, displays it, checks if it is already in ascending/descending order, and counts the number of odd and even numbers in it.

```python
def check_order(list):
    asc, des = sorted(list[:]), sorted(list[:])[::-1]
    if list == asc:
        print("List is already in ascending order")
    elif list == des:
        print("List is already in descending order")
    else:
        print("List is unsorted. The original list is", list)
        print("Its ascending order is:", asc)
        print("Its descending order is:", des)
```

```python
def count_odd_even(list):
    count_even, count_odd = 0, 0
    for num in list:
        if num%2 == 0:
            count_even += 1
        else:
            count_odd += 1
    print(f"In list {list}, the number of odd integers are {count_odd} and the number of even integers are {count_even}")
```

```python
def main():
    choice = 0
    while choice != 5:
        print("\n==================== MAIN MENU ====================")
        print("1. Input a list")
        print("2. Display the original list")
        print("3. Check if list is sorted")
        print("4. Count the number of odd and even elements")
        print("5. Quit")
        print("===================================================\n")
        
        try:
            choice = int(input("\nEnter your choice: "))
            if choice not in [1,2,3,4,5]: raise ValueError
        except:
            print("\nPlease input an integer from 1-4.")
        
        if choice == 1:
            n = int(input("Enter the number of integer elements: "))
            print(f"\nInput your {n} elemnts one-by-one below.")
            l = []
            for i in range(n):
                l.append(int(input(f"Input element {i+1}: ")))
        elif choice == 2:
            try:
                print("Your original list is:", l)
            except:
                print("You must input a list first.")
        elif choice == 3:
            try:
                check_order(l)
            except:
                print("You must input a list first.")
        elif choice == 4:
            try:
                count_odd_even(l)
            except:
                print("You must input a list first.")
        else:
            break
    print("End of program.")
```

```python
main()
```
