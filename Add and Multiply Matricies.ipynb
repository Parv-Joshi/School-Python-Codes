###### Exercise. Create a program that asks the user to input two matrices, display them, add and multiply them.

```python
def read(A, m, n):
    print("Enter", m * n, "elements one by one:")
    for i in range(m):
        for j in range(n):
            A[i][j] = eval(input(f"Element at row {i+1} and column {j+1}: "))
```

```python
def add(A, B, D, m, n, p):
    for i in range(m):
        for j in range(p):
            D[i][j] = A[i][j] + B[i][j]
    print("\nMatrix added.")
```

```python
def multiply(A, B, C, m, n, p):
    for i in range(m):
        for j in range(p):
            C[i][j] = 0
            for k in range(n):
                C[i][j] = A[i][k] * B[k][j]
    print("\nMatrix multipled.")
```

```python
def display(A, m, n):
    for i in range(m):
        for j in range(n):
            print(A[i][j], "\t", end=' ')
        print()
```

```python
def main():
    m = n = eval(input("Enter no. of rows/columns of 1st matrix: "))
    p = eval(input("Enter no. of rows/columns of 2st matrix: "))
    A = [[0 for col in range(n)] for row in range(m)]
    B = [[0 for col in range(p)] for row in range(n)]
    C = [[0 for col in range(p)] for row in range(m)]
    D = [[0 for col in range(p)] for row in range(m)]
    choice = 0
    while choice != 9:
        print("\n********* MAIN MENU **********")
        print("1. Enter Matrix 1")
        print("2. Display Matrix 1")
        print("3. Enter Matrix 2")
        print("4. Display Matrix 2")
        print("5. Add Matrix")
        print("6. Display Added Matrix")
        print("7. Multiply Matrix")
        print("8. Display Multiplied Matrix")
        print("9. Exit")
        print("******************************\n")
        choice = eval(input("Enter your choice: "))
        if choice == 1:
            read(A, m, n)
        elif choice == 2:
            display(A, m, n)
        elif choice == 3:
            read(B, n, p)
        elif choice == 4:
            display(B, n, p)
        elif choice == 5:
            add(A, B, D, m, n, p)
        elif choice == 6:
            display(D, m, n)
        elif choice == 7:
            multiply(A, B, C, m, n, p)
        elif choice == 8:
            display(C, m, p)
        elif choice == 9:
            print("Ending the program!")
```

```python
main()
```
