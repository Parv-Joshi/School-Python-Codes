###### Exercise. Create a program that asks the user to input a matrix, display is, calculates the sum of each row and column, finds the sum of major diagonal elements, minor elements, and all elements. Also, it must display the elements above and below the major and minor diagonal, the outline matrix, the middle row and column, the major and minor diagonals. Also, it must display the even elements only, find their count and sum. Also, it must display the odd elements only, find their count and sum. It must also display the transposed matrix, the product of both diagonal elements, and display swapped 1st and last row and column.

```python
n = eval(input("\nEnter no. of columns: "))
m = eval(input("\nEnter no. of rows: "))
A = [[0 for col in range(n)] for row in range(m)]
print("Enter", m * n, "elements of the matrix one-by-one")
for i in range(m):
    for j in range(n):
        A[i][j] = eval(input(f"Enter element at row {i+1} and column {j+1}: "))

print("\nThe Matrix is:")
for i in range(m):
    for j in range(n):
        print(A[i][j], "\t", end=' ')
    print()

print("\nRow wise sum:")
for i in range(m):
    row_sum = 0
    for j in range(n):
        row_sum += A[i][j]
    print("Sum of row", i + 1, "=", row_sum)

print("\nColumn wise sum:")
for j in range(n):
    column_sum = 0
    for i in range(m):
        column_sum += A[i][j]
    print("Sum of column", j + 1, "=", column_sum)

print("\nMajor Total:")
maj_total = 0
for j in range(n):
    for i in range(m):
        if i == j:
            maj_total += A[i][j]
print("Major Total =", maj_total)

print("\nMinor Total:")
min_total = 0
for j in range(n):
    for i in range(m):
        if i + j == n - 1:
            min_total += A[i][j]
print("Minor Total =", min_total)

print("\nTotal Sum:")
sum = 0
for i in range(m):
    for j in range(n):
        sum += A[i][j]
print("The total sum =", sum)

print("\nHighest and Lowest no.:")
high = low = A[0][0]
for i in range(m):
    for j in range(n):
        if A[i][j] > high:
            high = A[i][j]
        elif A[i][j] < low:
            low = A[i][j]
print("Highest Value =", high)
print("Lowest Value =", low)

print("\nAbove Major:")
for i in range(m):
    for j in range(n):
        if i < j:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nBelow Major:")
for i in range(m):
    for j in range(n):
        if i > j:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nAbove Minor:")
for i in range(m):
    for j in range(n):
        if i + j < n - 1:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nBelow Minor:")
for i in range(m):
    for j in range(n):
        if i + j > n - 1:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nOutline Matrix:")
for i in range(m):
    for j in range(n):
        if i == 0 or j == 0 or i == m - 1 or j == n - 1:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nDisplay Middle row:")
for i in range(m):
    for j in range(n):
        if i == (m - 1) / 2:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nDisplay Middle column:")
for i in range(m):
    for j in range(n):
        if j == (n - 1) / 2:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nDisplay Major Diagonal:")
for i in range(m):
    for j in range(n):
        if i == j:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nDisplay Minor Diagonal:")
for i in range(m):
    for j in range(n):
        if i + j == n - 1:
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()

print("\nPrinting no.of, sum of, and the even elements:")
even_count = 0
even_sum = 0
for i in range(m):
    for j in range(n):
        if A[i][j] % 2 == 0:
            even_count += 1
            even_sum += A[i][i]
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()
print("Sum of even elements: ", even_sum)
print("No. of even elements: ", even_count)

print("\nPrinting no.of, sum of, and the odd elements:")
odd_count = 0
odd_sum = 0
for i in range(m):
    for j in range(n):
        if A[i][j] % 2 != 0:
            odd_count += 1
            odd_sum += A[i][i]
            print(A[i][j], "\t", end=' ')
        else:
            print("\t", end=' ')
    print()
print("Sum of odd elements: ", odd_sum)
print("No. of odd elements: ", odd_count)

print("\nThe Transpose Matrix is:")
B = [[0 for col in range(m)] for row in range(n)]
for i in range(m):
    for j in range(n):
        print(B[i][j], "\t", end=' ')
    print()

print("\nProduct of Diagonals:")
major_product = 1
minor_product = 1
for i in range(m):
    for j in range(n):
        if i == j:
            major_product *= A[i][j]
        elif i + j == n - 1:
            minor_product *= A[i][j]
print("Product of Major Diagonal:", major_product)
print("Product of Minor Diagonal:", minor_product)

print("\nSwap Diagonals:")
for i in range(m):
    A[i][i], A[i][n - i - 1] = A[i][n - i - 1], A[i][i]
for i in range(m):
    for j in range(n):
        print(A[i][j], "\t", end=' ')
    print()

print("\nSwap 1st and last row:")
for i in range(m):
    A[i][0], A[i][n - 1] = A[i][n - 1], A[i][0]
for i in range(m):
    for j in range(n):
        print(A[i][j], "\t", end=' ')
    print()

print("\nSwap 1st and last row:")
for i in range(m):
    A[0][i], A[n - 1][i] = A[n - 1][i], A[0][i]
for i in range(m):
    for j in range(n):
        print(A[i][j], "\t", end=' ')
    print()
```
